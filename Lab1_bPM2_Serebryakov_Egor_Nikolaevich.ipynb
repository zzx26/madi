```python
import numpy as np

def mat_gen():
    while True:
        print('import height or blank')
        m = input()
        if m.isspace() or m =='':
            m = 50

        m = int(m)
        if m < 0:
            print('import natural value!')
        else:
            break

    while True:
        print('import lenght or blank')
        n = input()
        if n.isspace() or n =='':
            n = 50

        n = int(n)
        if n < 0:
            print('import natural value!')
        else:
            break

    while True:
        print('import min value or blank')
        min_limit = input()
        if min_limit.isspace() or min_limit =='':
            min_limit = -250

        min_limit = int(min_limit)

        print('import max value or blank')
        max_limit = input()
        if max_limit.isspace() or max_limit =='':
            max_limit = 1000 + 15

        max_limit = int(max_limit)
        if max_limit - min_limit < 0:
            print('the segment is degenerate')
        else:
            break 

    return np.random.randint(min_limit, max_limit+1, (m, n))

showcase_mat = mat_gen()
print(showcase_mat)
```

```python
#сортировка выбором

mat = mat_gen()
print('before sorting')
print(mat)

def vblbor(mat):
    for i in range(np.ma.size(mat, 0)):
        for j in range(np.ma.size(mat, 1)):
            mat[i, [j, np.argmin((mat[i, j:])) + j]] = mat[i, [np.argmin((mat[i, j:])) + j, j]]

            
print('after sorting')
vblbor(mat)
print(mat)
```

```python
#сортировка вставкой

mat = mat_gen()
print('before sorting')
print(mat)

def vstavka(mat):
    for i in range(np.ma.size(mat, 0)):
        for j in range(1, np.ma.size(mat, 1)):
            key = mat[i, j]
            k = 1
            while True:
                if k <= j and key < mat[i, j - k]:
                    key = 0
                    mat[i, j - k +1], mat[i, j - k] = mat[i, j - k], mat[i, j - k + 1]
                    key = mat[i, j - k]
                    k += 1
                else: 
                    break


print('after sorting')
vstavka(mat)
print(mat)            
    
```

```python
#турнирная

mat = mat_gen()
print('before sorting')
print(mat)

queue_lenght = 16

def tournament(arr):
    a = [i for i in arr]
    if len(a) <= queue_lenght:
        return np.array(small_tournament(a))
    else:
        return np.array(bracketsize(a))

def bracketsize(a):
    lenght = len(a)
    queue = []
    winners = []
    losers = []
    
    for i in range(lenght):
        queue.append(a[i])
        a.pop(i)
        
    while len(a) > 0:
        if not winners:
            winners.append(queue[0])
            queue.pop(0)
            
        if a[0] > winners[len(winners)]:
            queue.append(a[0])
            a.pop(0)
        else:
            losers.append(a[0])
            a.pop(0)
        
        if queue:
            winners.append(queue[0])
            queue.pop(0)
            
    while queue:
        winners.append(queue[0])
        queue.pop(0)
    if not losers:
        return winners
    else:
        a = losers + winners
        while len(a) > lenght:
            a.pop()
        bracketsize(a)
    
def small_tournament(a):
    arr = np.sort(np.array(a))
    a = [i for i in arr]


    
def tournament_2d(mat):
    for i in range(np.ma.size(mat, 0)):
        tournament(mat[i, :])
        
tournament_2d(mat)
print('after sorting')
print(mat)
```

```python
#быстрая

mat = mat_gen()
print('before sorting')
print(mat)

def quick(mat):
    def quicksort(a):
        l_a = len(a)
        if len(np.unique(a)) > 1:
            root = a[np.random.randint(0, l_a)]
            b = np.ones(len(a))
            for i in range(len(a)):
                if a[i] < root:
                    b[i] = -1
                elif a[i] == root:
                    b[i] == 0
            c_l = []
            c_e = []
            c_g = []
            for i in range(len(a)):
                if not b[i]:
                    c_e.append(a[i])
                elif b[i] == 1:
                    c_g.append(a[i])
                else:
                    c_l.append(a[i])
            c_1 = np.array(c_l)
            c_2 = np.array(c_e)
            c_3 = np.array(c_g)
            return np.concatenate((quicksort(c_1),  quicksort(c_2), quicksort(c_3)))
        else:
            return(a)


    for i in range(np.ma.size(mat, 0)):
        mat[i, :] = quicksort(mat[i, :])


print('after sorting')
quick(mat)
print(mat)

```

```python
#пирамидальная

mat = mat_gen()
print('before sorting')
print(mat)

def heapsort(mat):
    def heapify(a, lenght, i):
        root_index = i
        left_index = 2*root_index + 1
        right_index = 2*root_index + 2

        if left_index < lenght and a[root_index] < a[left_index]:
            root_index = left_index

        if right_index < lenght and a[root_index] < a[right_index]:
            root_index = right_index

        if root_index != i:
            a[i], a[root_index] = a[root_index], a[i]

            heapify(a, lenght, root_index)


    def pyrsort(a):
        n = len(a)

        for i in range(n//2 - 1 , -1, -1):
            heapify(a, n, i)

        for i in range(n - 1, 0, -1):
            a[i], a[0] = a[0], a[i]
            heapify(a, i, 0)


    for i in range(np.ma.size(mat, 0)):
        pyrsort(mat[i, :])

print('after sorting')
heapsort(mat)
print(mat)
```
